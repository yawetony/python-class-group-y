```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
import sklearn.preprocessing
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('bank.csv')
df.head(10)
```

```python
#The describe() function returns a dataframe with all the descriptive statistics of the dataframe in question. 
#Descriptive statistics, hence all the numerical variables are considered.

df.describe()
```

```python
#df.columns method is used to display the column names of the dataframe...
print(list(df.columns))
```

```python
#Import the seaborn library to visualise the relationship between different numerical variables. It is not required for this project
import seaborn as sns
sns.pairplot(df[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'y']])
```

```python
df.columns
```

```python
#We drop certain columns that we deem not necessary for the logistic regression process.
df_new = df.drop(df.columns[[0, 3, 7, 8, 9, 10, 11, 12, 13, 14]], axis =1)
```

```python
#The trimmed dataframe is saved to df_new, a new dataframe...

df_new
```

```python
#drop a few columns which are similar to NA valued columns.
data = pd.get_dummies(df_new, columns=['job','marital','default','housing','poutcome'],drop_first=False)
data.columns[[13,22,24]]
len(data.columns)
data = data.drop(data.columns[[13,22,24]],axis=1)
len(data.columns)
```

```python
#we must separate the dataframe to use into predictors and target. Y is the target and X is the predictor.

Y = data.iloc[:,1]
#y = data.iloc[:,1:2]
Y.head()

#X is the predictor.
X = data.drop(data.columns[[1]],axis=1)

#A glimpse of X...
X.head()
```

```python
#We do a split of the dataset using the train_test_split() function of sklearn.model_selection.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0)
```

```python
#The model specification and fit.
classifier = LogisticRegression(solver = "lbfgs", random_state= 0)
classifier.fit(X_train, Y_train)

#Predicting the model on the test set...
predicted_y = classifier.predict(X_test)

#Finding the accuracy of predictions by comparing the predicted values to the actual values of the test data...
sum(predicted_y.tolist() == Y_test)/len(Y_test)
```

```python
#The classifier.score() function gives a score of 0.89 which means that our model predicts with an accuracy of 89%. This is just the accuracy metric.

classifier.score(X_test, Y_test)
```
